```python
import sys
sys.path.append('../python/')
import NGC5533_functions as nf
import matplotlib.pyplot as plt
import numpy as np
from scipy.interpolate import InterpolatedUnivariateSpline
```

```python
#Bulge

r = np.linspace(0.1,20,10)
#b_c = 1
#print(nf.b_vsquarev([1,2],2.5)**(1/2))
#vbspline = InterpolatedUnivariateSpline(r, b_c*nf.b_v(r), k=3)    #turns array vb into an approximate function
vbspline = InterpolatedUnivariateSpline(r, nf.b_v(r), k=3)
#print(vbspline.get_coeffs())
```

```python
# Bulge spline plot

fig = plt.figure(figsize=(9.0,8.0))

plt.plot(r,vbspline(r),label='Bulge Spline')

plt.legend()
plt.show()
```

```python
# Disk

#r = np.linspace(0.1,20,10)
pref = 1
#d_c = 1
b = nf.d_v(r)
b[np.isnan(b)] = 0
vdspline = InterpolatedUnivariateSpline(r, b, k=3)    #turns array vb into an approximate function
```

```python
# Black Hole

r = np.linspace(0.1,20,10)
#bh_c = 1
vbhspline = InterpolatedUnivariateSpline(r, nf.bh_v(r), k=3)    #turns array vb into an approximate function
```

```python
# Halo

#r = np.linspace(0.1,20,10)
#h_c = 1
vhalospline = InterpolatedUnivariateSpline(r, nf.h_v(r), k=3)    #turns array vb into an approximate function
```

```python
# Spline plot

fig = plt.figure(figsize=(9.0,8.0))

plt.plot(r,vbspline(r),label='Bulge Spline')
plt.plot(r,vdspline(r),label='Disk Spline')
plt.plot(r,vbhspline(r),label='Black Hole Spline')
plt.plot(r,vhalospline(r),label='Halo Spline')

plt.legend()
plt.show()
```
