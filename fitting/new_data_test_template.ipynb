```python
import sys

sys.path.append('../python/')
import dataPython as dp
import numpy as np
import scipy.interpolate as inter

#data = dp.getXYdata('data/graph.txt')
data = dp.getXYdata_wYerr('data/M31_data.txt')
r_dat = np.asarray(data['xx'])
v_dat = np.asarray(data['yy'])
#v_err0 = np.asarray(data['ex'])
v_err1 = np.asarray(data['ey'])

#plotting
import matplotlib as mpl
from matplotlib import pyplot as plt
import pylab

#plotting
fig = plt.figure(figsize=(20.0,16.0))                 #size of the plot
ax3 = plt.axes()
ax3.grid(True)
ax3.yaxis.grid(True,which='minor',linestyle='--')

#xmax = 19

#plt.scatter(r_dat,v_dat)
plt.errorbar(r_dat,v_dat,yerr=v_err1, marker='o', markersize=8, \
             ecolor='gray',color='gray', linestyle='none', label='NGC 5533 (Noordermeer 2007)', linewidth=2)
    
#plt.axis('tight')

plt.xlim(0,30)
plt.ylim(0,300)
ax3.set_title('Galaxy Name', fontsize=18)              #labeling axes
ax3.set_xlabel('Radius (kpc)', fontsize=14)
ax3.set_ylabel('Velocity (km/s)', fontsize=14)
ax3.xaxis.set_tick_params(labelsize=14)
ax3.yaxis.set_tick_params(labelsize=14);
#plt.show()
```
