```python
#Dependencies
import sys
sys.path.append('python')
import NGC5533_functions as nf
import dataPython as dp
import matplotlib.pyplot as plt
import numpy as np
```

```python
#Data points
gdata = dp.getXYdata('fitting/data/final/nord-120kpc-disk-gipsy.txt') #data from GIPSY
ndata = dp.getXYdata('fitting/data/final/nord-120kpc-disk-gipsy-mass80.txt') #testing with a different mass value
rdat = np.asarray(gdata['xx']) #radius values
gvel = np.asarray(gdata['yy']) #velocities from GIPSY
nvel = np.asarray(ndata['yy']) #new velocities
cvel = nf.d_v(rdat,pref=0.4,load=True,path='fitting')  #calculated velocities
```

```python
#Plotting
plt.plot(rdat,gvel,label='Mass 160')
plt.plot(rdat,cvel,label='Calculated')
plt.plot(rdat,nvel,label='Mass 80')
plt.legend()
plt.show()
```
