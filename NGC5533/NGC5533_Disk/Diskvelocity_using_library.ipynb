```python
#Imports
import sys
sys.path.append('../python/')
import NGC5533_functions as nf

import numpy as np
import matplotlib.pyplot as plt
```

```python
fig = plt.figure(figsize=(9.0,8.0))

plt.xlim(0,120)
#plt.ylim(0,450)

r = np.linspace(0.1,120,100)
pref=0.3

def disk(r,pref,h,drho00):
    return nf.d_v(r,pref,nf.h_c,nf.drho00_c, load=True)
def halo(r,rc,hrho00):
    return nf.h_v(r,nf.h_rc,nf.hrho00_c, load=True)

def diskminushalo(r,pref,h,drho00,rc,hrho00):
    return np.sqrt((disk(r,pref,h,drho00)**2)-(halo(r,rc,hrho00)**2))

def f(r):
    return 1410 / np.sqrt(r)

rplot = np.linspace(37.38,120,100)
plt.plot(rplot, f(rplot),'purple',label='1/sqrt(r)')
plt.plot(r,nf.d_v(r,pref,nf.h_c,nf.drho00_c,load=True),label='Disk: from library ')
#plt.plot(r,diskminushalo(r,pref,nf.h_c,nf.drho00_c,nf.h_rc,nf.hrho00_c), label='Disk minus Halo')
#plt.plot(r,nf.d_v(r,pref=1,nf.h_c,nf.drho00_c),label='Disk: from library (innerintegral: 0 to 125)')
#plt.plot(r,nft.d_v(r,pref=1,nf.h_c,nf.drho00_c),label='Disk: from test library (innerintegral: 0 to inf)')
#plt.plot(r,nf.b_v(r,nf.n_c,nf.re_c,load=True),label='Bulge: from library')
#plt.plot(r,nf.h_v(r,nf.h_rc,nf.hrho00_c,load=True),label='Halo: from library')

plt.legend()
plt.show()
```
