```python
import numpy as np
import matplotlib as mpl
from matplotlib import pyplot as plt

fig = plt.figure(figsize=(9.0,8.0))                 #size of the plot
ax1 = plt.axes()
ax1.grid(True)
ax1.yaxis.grid(True,which='minor',linestyle='--')

 #import text files

# n = 0.5
x05 = []
y05 = []
for line in open('data/n05_bulge_v.txt', 'r'):
    values = [float(s) for s in line.split()]
    x05.append(values[0])
    y05.append(values[1])

# n = 1.0
x1 = []
y1 = []
for line in open('data/n1_bulge_v.txt', 'r'):
    values = [float(s) for s in line.split()]
    x1.append(values[0])
    y1.append(values[1])

# n = 2.0
x2 = []
y2 = []
for line in open('data/n2_bulge_v.txt', 'r'):
    values = [float(s) for s in line.split()]
    x2.append(values[0])
    y2.append(values[1])

# n = 3.0
x3 = []
y3 = []
for line in open('data/n3_bulge_v.txt', 'r'):
    values = [float(s) for s in line.split()]
    x3.append(values[0])
    y3.append(values[1])
    
# n = 4.0
x4 = []
y4 = []
for line in open('data/n4_bulge_v.txt', 'r'):
    values = [float(s) for s in line.split()]
    x4.append(values[0])
    y4.append(values[1])

# n = 6.0
x6 = []
y6 = []
for line in open('data/n6_bulge_v.txt', 'r'):
    values = [float(s) for s in line.split()]
    x6.append(values[0])
    y6.append(values[1])  
    
 #curves
plt.plot(x05, y05, ':b', linestyle='solid', label='n = 0.5')
plt.plot(x1, y1, ':g', linestyle='solid', label='n = 1.0')
plt.plot(x2, y2, ':r', linestyle='solid', label='n = 2.0')
plt.plot(x3, y3, ':c', linestyle='solid', label='n = 3.0')
plt.plot(x4, y4, ':m', linestyle='solid', label='n = 4.0')
plt.plot(x6, y6, ':y', linestyle='solid', label='n = 6.0')

plt.axis([0, 10, 0, 160]);                         #xmin, xmax, ymin, ymax
                                                   #or use: plt.xlim(10, 0) plt.ylim(1.2, -1.2);
                                                   #or to tighten the bounds: plt.axis('tight')
                                                   #or use: plt.axis('tight')
                                                   #plt.axis([0, 15000, 0.5e6, 2.5e6])
 #labeling plots
#ax3.set_ylim(0,np.max(v)*1.3)
ax1.set_title('Bulge Rotation Curve', fontsize=18)              #labeling axes
ax1.set_xlabel('Radius (kpc)', fontsize=14)
ax1.set_ylabel('Velocity (km/s)', fontsize=14)
ax1.xaxis.set_tick_params(labelsize=14)
ax1.yaxis.set_tick_params(labelsize=14);

plt.legend(fontsize=16);                                   #adding legend
plt.tight_layout()
plt.savefig('Plot1.eps')
plt.show()
```
