```python
import numpy as np
import matplotlib as mpl
from matplotlib import pyplot as plt
import scipy.integrate as integrate
from scipy.integrate import quad
import scipy.optimize as so
import scipy.special as ss
import math

h = 4.9                                                     #radial scalelength in kpc
z0 = .99                                                    #half-thickness in kpc 
L0 = 5.0e9                                                  #units? value?

x = np.linspace(10, 50, 200)


Lrz = lambda r,z: L0*np.exp(-r/h)*(math.cosh(z/z0))**-2     #cosh in radians?

y = np.zeros(np.shape(x))

print(y)
for r,z in enumerate(x):
    hr = lambda r: Lrz(r,z)
    yr = quad(hr, 1, r)[0]
    y[r] = yr
print(y)
```

```python
v = np.sqrt(y)

fig = plt.figure(figsize=(9.0,8.0))                 #size of the plot
ax3 = plt.axes()
ax3.grid(True)
ax3.yaxis.grid(True,which='minor',linestyle='--')


 #functions
plt.plot(x, v, linestyle='-.', label='Velocity(r)')          

plt.axis('tight');                                 #xmin, xmax, ymin, ymax
                                                   #or use: plt.xlim(10, 0) plt.ylim(1.2, -1.2);
                                                   #or to tighten the bounds: plt.axis('tight')
                                                   #or use: plt.axis('tight')
                                                   #plt.axis([0, 15000, 0.5e6, 2.5e6])
 #labeling plots
ax3.set_title('Luminosity Profile for Disk', fontsize=25)              #labeling axes
ax3.set_xlabel('Radius (kpc)', fontsize=14)
ax3.set_ylabel('Luminosity ($L_\odot$)', fontsize=25)

plt.legend(fontsize=14);                                   #adding legend
plt.tight_layout()
plt.savefig('Luminosity Profile for Disk.png')
plt.show()
```
