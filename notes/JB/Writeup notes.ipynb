## Bulge

>__Luminosity__: <br>
    \begin{equation}
    I_b(\kappa) = I_0 \exp \left[-\left(\frac{\kappa}{r_0} \right)^{\frac{1}{n}}\right]
    \end{equation}
    <br>
>>where:<br>
        $I_0$ = central surface brightness<br>
        $n$ = concentration parameter<br> 
        $r_0$ = characteristic radius (related to $r_e$)<br>
        $r_e$ = effective radius<br>
<br>
>>Gamma function: <br>
from Noordermeer's paper<br>
    \begin{equation}
    \gamma(b_n,2n) = \frac{1}{2} \Gamma(2n)
    \end{equation}<br>
    \begin{equation}
    r_e = b_n^n r_0 b_n 
    \end{equation}<br>
from Anthony's calculations<br>
    \begin{equation}
    \gamma(n,u_e) = \frac{1}{2} \Gamma(n)
    \end{equation}<br>

>__Velocity__: <br>
    <font color=green>
    \begin{equation}
V_{B}^{2}(r) = C \int_{m=0}^{r} \left[\int_{\kappa=m}^{\infty} \frac {e^{-\left(\frac{\kappa}{r_0}\right)^{\frac{1}{n}}} \left(\frac{\kappa}{r_0}\right)^{\frac{1}{n-1}}} {\sqrt{\kappa^2-m^2}} d\kappa\right] \times \frac{m^{2}}{\sqrt{r^{2}-m^{2}e^{2}}} dm
    \end{equation}<br></font>
>>where:<br>
    \begin{equation}
    C = \frac{4Gq\Upsilon_b I_0}{r_0n} \sqrt{sin^{2}i + \frac{1}{q^{2}} cos^{2}i}
    \end{equation}<br>
    \begin{equation}
    e = \sqrt{1-q^{2}}
    \end{equation}<br> 
        $\Upsilon_b$ = mass-to-light ratio<br>
        $e$ = eccentricity of the bulge<br>
        $q$ = intrinsic axis ratio<br> 
        $i$ = inclination angle<br> 
        $G$ = gravitational constant<br>
    <br>

    
- Parameters for NGC5533 <br>
> $n$ = 2.7<br>
> $r_e$ = 2.6 [kpc] <br>
> $I_0$ = 4.10e10<br>
> $q$ = 0.33 <br>
> $\Upsilon_b$ = 2.8   $[M_\bigodot / L_\bigodot]$<br>
> $i$ = 45&deg; <br>
> $G$ = 4.300e-6   [kpc($M_\bigodot)^{-1}$$(km/s)^{2}$]
- References <br>
>Noordermeer, Edo. __"The rotation curves of flattened Sérsic bulges."__ _Monthly Notices of the Royal Astronomical Society_ 385, no. 3 (2007): 1359-1364.

## Disk
>__Density Distribution__: <br>
    \begin{equation}
    \rho(r,z) = \rho_0(r)\left[\cosh\left(\frac{z}{z_0}\right)\right]^{-2}
    \end{equation}<br>
    with
    \begin{equation}
    \rho_0(r) = 
    \begin{cases}
    \rho_{00}(r)e^{-\frac{r}{h}}       & \quad \text{if } r\leq R \\
    \rho_{0}(R)\left[1-\frac{r-R}{\delta}\right]       & \quad \text{if } R<r\leq R+\delta \\
    0       & \quad \text{if } r > R+\delta \\
  \end{cases}
\end{equation}
    <br>
>>where:<br>
    $\rho_{00}$ = central density<br>
    $h$ = radial scale length<br>
    $R$ = cut-off radius<br>
    $\delta$ = cut-off length upper limits<br>
    $z_0$ = half-thickness
>>\begin{equation}
   x = \frac{r^{2}+u^{2}+\xi^{2}}{2ru}
   \end{equation}<br>
   \begin{equation}
   p = x - \sqrt{x^{2}-1}
   \end{equation}<br>
   
>__Mass of Disk__: <br>
   \begin{equation}
   M_d = \int_0^{\infty} rdr \int_{-\infty}^{\infty} dz \rho(r,z).
   \end{equation}<br>
>>Prefactor
   \begin{equation}
   Pref = \Upsilon_d \cdot \frac{L_{d,R}}{\int_0^{\infty} rdr \int_{-\infty}^{\infty} dz \rho(r,z)}
   \end{equation}<br>
>>Absolute Magnitude to Luminosity
   \begin{equation}
   L_0 = 10^{0.4\left(Mag_{sun} - Mag_{abs}\right)}
   \end{equation}<br>
>>where:<br>
    $L_0$ = Luminosity<br>
    $Mag_{sun}$ = Absolute Magnitude of Sun<br>
    $Mag_{abs}$ = Absolute Magnitude<br>

>__Gravitational Force__: <br>
    \begin{equation}
    F_{rd}(r) = 4\pi G \int_{0}^{\infty} u  du \int_{0}^{\infty} dz \left[\frac{\partial\rho(u,z)}{\partial u}\right] 2 \frac{\mathscr{K}(p)-\mathscr{E}(p)}{\pi\sqrt{rup}} 
    \end{equation}<br>
>>where:<br>
        $\mathscr{K}$ and $\mathscr{E}$ = complete elliptic integrals<br>
        $G$ = gravitational constant<br>
    
>__Velocity__: <br>
    <font color=green>
    \begin{equation}
    V_{D}^{2}(r) = -rF_{rd}(r)
    \end{equation}<br></font>
   
- Parameters for NGC5533 <br>

> $\delta$ = 0.2h [kpc]<br>

<br>

## Dark Halo
>__Isothermal Density Profile__: <br>
    \begin{equation}
    \rho(r) = \rho_{00}\left[1+{\left(\frac{r}{r_c}\right)}^{2}\right]^{-1}
    \end{equation}
    <br>
>>where:<br>
    $\rho_{00}$ = central density<br>
    $r_c$ = core radius<br>
    
>__Mass__: <br>
    \begin{equation}
    M_{DM}(r) = \int_{0}^{r} 4\pi\rho(R)R^{2}dR
    \end{equation}<br>
>>where:<br>
        $M_{DM}$ = mass of the dark halo<br>
    <br>
    
>__Velocity__: <br>
    <font color=green>
    \begin{equation}
    V_{DM}^{2}(r) = \frac{GM_{DM}(r)}{r}
    \end{equation}<br></font>
>>where:<br>
        $G$ = gravitational constant<br>
    <br>
    
- Parameters for NGC5533 <br>
> $\rho_{00}$ = 0.31e9 [$M_\bigodot$ /kpc^3]<br>
> $r_c$ = 1.4 [kpc]<br>
> $G$ = 4.300e-6   [kpc($M_\bigodot)^{-1}$$(km/s)^{2}$]<br>

- References <br>
>Jimenez, Raul, Licia Verde, and S. Peng Oh. __"Dark halo properties from rotation curves."__ _Monthly Notices of the Royal Astronomical Society_ 339, no. 1 (2003): 243-259.<br><br>
>Karukes, E. V., P. Salucci, and Gianfranco Gentile. __"The dark matter distribution in the spiral NGC 3198 out to 0.22 $R_{vir}$."__ _Astronomy & Astrophysics_ 578 (2015): A13.
