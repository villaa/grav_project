```python
import numpy as np
import matplotlib as mpl
from matplotlib import pyplot as plt
import scipy.integrate as integrate
from scipy.integrate import quad
from scipy.special import ellipe
from scipy.special import ellipk

#halo density function
p0=3.0                                                #finite surface density
rc=4.0                                                #break radius
pr = lambda r: p0/(1 + (r/rc)**2)**-1

print(pr(2))
```

```python
#rotation curve for halo

G = 4.300e-6                                                #gravitational constant (kpc/solar mass*(km/s)^2)
z = 1.0                                                     #thickness of galactic plane?
vh = lambda r: np.sqrt(2*np.pi*G*np.sqrt(r**2+z**2))        #from casertano eq 9 solved for v
print(vh(0))
```

```python
#plot halo velocity
fig = plt.figure(figsize=(9.0,8.0))                #size of the plot
ax = plt.axes()
ax.grid(True)
ax.yaxis.grid(True,which='minor',linestyle='--')

r = np.linspace(0, 20, 20)
plt.plot(r, vh(r), linestyle='solid', label='Disk Velocity')          

plt.axis('tight');                                 #xmin, xmax, ymin, ymax
                                                   #or use: plt.xlim(10, 0) plt.ylim(1.2, -1.2);
                                                   #or to tighten the bounds: plt.axis('tight')
                                                   #or use: plt.axis([0, 15000, 0.5e6, 2.5e6])

fig.suptitle('Halo Velocity', fontsize=18)
ax.set(title='Halo Component NGC 5533')              #labeling axes
ax.set_xlim(0, 20)
ax.set_xlabel('r (kpc)', fontsize=14)
ax.set_ylabel('v (km/s)', fontsize=14)
ax.xaxis.set_tick_params(labelsize=14)
ax.yaxis.set_tick_params(labelsize=14);

plt.legend(fontsize=16);                                   #adding legend
plt.savefig('Halo Velocity200pts.png')
plt.show()
```
